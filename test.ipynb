```python


def create_dataframe(savedTracks):
    """Creates a DataFrame from a list of saved tracks with additional fields."""
    columns = [
        'added_at', 'track_name', 'album_name', 'artist_names', 'album_release_date',
        'duration_ms', 'explicit', 'popularity', 'track_id', 'album_id', 'artist_ids', 'isrc'
    ]
    df = pd.DataFrame(columns=columns)

    for x in range(len(savedTracks['items'])):
        row = savedTracks['items'][x]
        track = row['track']
        album = track['album']
        #print("album",album)
        artists = track['artists']

        added_at = row['added_at']
        track_name = track['name']
        album_name = album['name']
        artist_names = ', '.join([artist['name'] for artist in artists]) # Join artist names
        album_release_date = album['release_date']
        duration_ms = track['duration_ms']
        explicit = track['explicit']
        popularity = track['popularity']
        track_id = track['id']
        album_id = album['id']
        artist_ids = ', '.join([artist['id'] for artist in artists])
        isrc = track['external_ids'].get('isrc') # Handle missing isrc

        df.loc[len(df)] = [
            added_at, track_name, album_name, artist_names, album_release_date,
            duration_ms, explicit, popularity, track_id, album_id, artist_ids, isrc
        ]

    return df
```

```python
import spotipy
import pandas as pd
from collections import Counter

def get_artist_data(spotify_client, artist_ids):
    """Retrieves artist data from Spotify API."""

    artist_data = {}
    for i in range(0, len(artist_ids), 100):
        batch_ids = artist_ids[i:i + 100]
        results = spotify_client.artists(batch_ids)
        for artist in results['artists']:
            artist_data[artist['id']] = artist
    return artist_data

def enrich_track_dataframe(spotify_client, track_df):
    """Enriches the track DataFrame with artist data."""

    track_artist_ids = set()
    album_artist_ids = set()

    for row in track_df.itertuples(index=False):
        if row.track_artist_ids:
            track_artist_ids.update(row.track_artist_ids.split(', '))
        if row.album_artist_ids:
            album_artist_ids.update(row.album_artist_ids.split(', '))

    all_artist_ids = list(track_artist_ids.union(album_artist_ids))

    if not all_artist_ids:
        return track_df

    artist_data = get_artist_data(spotify_client, all_artist_ids)

    new_columns = [
        'artist_genres', 'artist_popularity',
        'album_artist_genres', 'album_artist_popularity'
    ]
    for col in new_columns:
        track_df[col] = None

    for index, row in track_df.iterrows():
        track_artists = row['track_artist_ids'].split(', ') if row['track_artist_ids'] else []
        album_artists = row['album_artist_ids'].split(', ') if row['album_artist_ids'] else []

        track_genres = []
        track_artist_pops = []
        album_genres = []
        album_artist_pops = []

        for artist_id in track_artists:
            artist = artist_data.get(artist_id)
            if artist:
                track_genres.extend(artist.get('genres', []))
                track_artist_pops.append(artist.get('popularity', 0))

        for artist_id in album_artists:
            artist = artist_data.get(artist_id)
            if artist:
                album_genres.extend(artist.get('genres', []))
                album_artist_pops.append(artist.get('popularity', 0))

        track_df.at[index, 'artist_genres'] = ', '.join(track_genres)
        track_df.at[index, 'artist_popularity'] = sum(track_artist_pops) / len(track_artist_pops) if track_artist_pops else 0
        track_df.at[index, 'album_artist_genres'] = ', '.join(album_genres)
        track_df.at[index, 'album_artist_popularity'] = sum(album_artist_pops) / len(album_artist_pops) if album_artist_pops else 0

    return track_df

def analyze_genres(track_df):
    """Analyzes genres from the track DataFrame."""

    all_genres = []
    for genres in track_df['artist_genres'].dropna():
        all_genres.extend(genres.split(', '))
    for genres in track_df['album_artist_genres'].dropna():
        all_genres.extend(genres.split(', '))

    genre_counts = Counter(all_genres)
    genre_df = pd.DataFrame(list(genre_counts.items()), columns=['Genre', 'Track Count'])

    genre_popularity = track_df.groupby(['artist_genres'])['artist_popularity'].mean().reset_index()
    genre_popularity.rename(columns={'artist_genres': 'Genre', 'artist_popularity': 'Average Artist Popularity'}, inplace=True)

    song_genre_popularity = track_df.groupby(['artist_genres'])['track_popularity'].mean().reset_index()
    song_genre_popularity.rename(columns={'artist_genres': 'Genre', 'track_popularity': 'Average Song Popularity'}, inplace=True)

    genre_df = pd.merge(genre_df, genre_popularity, on='Genre', how='left')
    genre_df = pd.merge(genre_df, song_genre_popularity, on='Genre', how='left')

    return genre_df

# Example Usage
enriched_df = enrich_track_dataframe(spotifyClient, all_tracks_df)
genre_analysis_df = analyze_genres(enriched_df)
print(genre_analysis_df)
```
